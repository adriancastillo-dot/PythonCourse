```python
class Rectangulo:
	def __init__(self, base, altura):
		self.base = base
		self.altura = altura
		
	def perimetro(self):
		return 2*(self.base + self.altura)
	
	def area(self):
		return self.base * self.altura
		
r1 = Rectangulo(2,3)
print("Perimetro: ", r1.perimetro())
print("Area: ", r1.area())
```

```python
class Coche:
	def __init__(self, marca, modelo):
		self.marca = marca
		self.modelo = modelo
		self._velocidad = 0
	
	def acelerar(self):
		self._velocidad += 10
		print(f"{self.marca} {self.modelo} acelera a {self._velocidad} km/h.")
		
	def frenar(self):
		self._velocidad -= 10
		print(f"{self.marca} {self.modelo} frena a {self._velocidad} km/h.")
	
c1 = Coche("Peugeot","208")
c1.acelerar()
c1.acelerar()
c1.frenar()
```

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo
    
    def depositar(self, cantidad):
        self.saldo += cantidad

    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
        else:
            print("Saldo insuficiente")
    
    def mostrar_info(self):
        print(f"Titular: {self.titular}, Saldo: {self.saldo} EUR")

c = CuentaBancaria("Adri", 1000)
c.depositar(150)
c.retirar(300)
c.mostrar_info()    
```

```python
class Animal:
	def moverse(self):
		pass
	
class Ave(Animal):
	def moverse(self):
		print("El ave se desplaza por el aire")
		
class Perro(Animal):
	def moverse(self):
		print("El perro se desplaza por tierra")
		
class Pez(Animal):
	def moverse(self):
		print("El pez se desplaza por el mar")

def hacer_moverse(animal):
	animal.moverse()
	
animales = [Ave(), Perro(), Pez()]

for i in animales:
	hacer_moverse(i)
```

```python
from abc import ABC, abstractmethod

class Vehiculo(ABC):
	@abstractmethod
	def arrancar(self):
		pass

class Coche(Vehiculo):
	def arrancar(self):
		print("El coche arrancó")
		
class Moto(Vehiculo):
	def arrancar(self):
		print("La moto arrancó")

vehiculos = [Coche(), Moto()]

for i in vehiculos:
	i.arrancar()
```

```python
class Cuenta:
    def __init__(self, titular, saldo_inicial=0.0):
        # Guardamos titular y saldo como atributos PRIVADOS
        self.__titular = titular
        self.__saldo = saldo_inicial

    def depositar(self, cantidad):
        """Añade cantidad al saldo si es positiva."""
        if cantidad > 0:
            self.__saldo += cantidad
        else:
            print("Entrada inválida: la cantidad debe ser positiva.")

    def retirar(self, cantidad):
        """Resta cantidad del saldo si hay suficiente dinero."""
        if cantidad > self.__saldo:
            print("Saldo insuficiente.")
        elif cantidad <= 0:
            print("La cantidad debe ser positiva.")
        else:
            self.__saldo -= cantidad

    def mostrar_info(self):
        """Muestra el titular y el saldo actual."""
        print(f"Titular: {self.__titular} | Saldo actual: {self.__saldo:.2f} €")


def main():
    # Probamos la clase
    c1 = Cuenta("Adrián", 1000)
    c1.retirar(300)
    c1.depositar(150)
    c1.mostrar_info()

    c2 = Cuenta("Marta", 50)
    c2.retirar(100)
    c2.depositar(-20)
    c2.mostrar_info()
    
if __name__ == "__main__":
    main()
```

```python
class Empleado:
    def __init__(self, nombre, sueldo):
        """Constructor de Empleado."""
        self.nombre = nombre
        self.sueldo = sueldo

    def mostrar_info(self):
        """Muestra la información básica del empleado."""
        print(f"Empleado: {self.nombre} - {self.sueldo} €")


class Gerente(Empleado):
    def __init__(self, nombre, sueldo, departamento):
        """Constructor de Gerente, hereda de Empleado."""
        super().__init__(nombre, sueldo)
        self.departamento = departamento

    def mostrar_info(self):
        """Muestra la información del gerente."""
        print(f"Gerente: {self.nombre} - {self.sueldo} € - {self.departamento}")


def main():
    e = Empleado("Ana", 2500)
    g = Gerente("Luis", 4000, "Ventas")

    e.mostrar_info()
    g.mostrar_info()


if __name__ == "__main__":
    main()

```

```python
from abc import ABC, abstractmethod
import math

class Figura(ABC):
    @abstractmethod
    def area(self):
        """Devuelve el área de la figura."""
        pass


class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        """Calcula el área del rectángulo."""
        return self.base * self.altura


class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        """Calcula el área del círculo."""
        return math.pi * (self.radio ** 2)


def main():
    # Crear varias figuras
    figuras = [
        Rectangulo(2, 3),
        Circulo(1),
        Circulo(2)
    ]

    # Recorrer la lista e imprimir el área de cada una
    for i in figuras:
        print(f"Área de {i.__class__.__name__}: {i.area():.2f}")


if __name__ == "__main__":
    main()
```
