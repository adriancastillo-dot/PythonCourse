```python
def process_text(text="  This is a DEFAULT text with Python and PYTHON words.  ",
                 words_to_replace=None):
    """
    Devuelve una tupla (nuevo_texto, num_reemplazos)
    """
    cont = 0

    # 1 Quitamos los espacios al principio y al final del texto y lo pasamos todo a minúsculas
    text = text.strip().lower()

    if words_to_replace is None:
        words_to_replace = ["python", "default"]

    #  Recorremos cada palabra de la lista de censura
    for word in words_to_replace:
        # Contamos cuántas veces aparece esa palabra (en minúsculas) en el texto actual
        reemplazos = text.count(word.lower())

        if reemplazos > 0:
            # Reemplazamos todas las apariciones de la palabra por asteriscos del mismo tamaño
            text = text.replace(word.lower(), '*' * len(word))
            cont += reemplazos

    #  Devolvemos una tupla con el texto final y el total de reemplazos
    return text, cont


def main():

    text = input("Introduce el texto a procesar: ").strip()
    censura = input("Introduce las palabras a censurar, separadas por comas (deja vacío para defecto): ").strip()

    #  Si el usuario ha escrito algo, separamos las palabras por comas
    #  y quitamos espacios sobrantes con strip(), descartando entradas vacías
    words_to_replace = [w.strip() for w in censura.split(",") if w.strip()] if censura else None

    #  Llamamos a la función process_text con los valores introducidos
    nuevo_texto, num_reemplazos = process_text(text, words_to_replace)

    print("\nTexto procesado:", nuevo_texto)
    print("Número de reemplazos:", num_reemplazos)

if __name__ == "__main__":
    main()

```

```python
tasks = []  # lista global


def add_task(task):
    """Añade una nueva tarea a la lista global tasks."""
    # TODO: usar la variable global tasks
    tasks.append(task)


def remove_task(task):
    """Elimina la primera aparición de task en tasks si existe."""
    # TODO: comprobar si task está y eliminarla
    if task in tasks:
        tasks.remove(task)
    else :
        print("La tarea no existe.")


def list_tasks():
    """Muestra todas las tareas numeradas."""
    # TODO: usar enumerate para numerar
    for i, task in enumerate(tasks, start=1):
        print("\n"f"{i}. {task}")


def task_manager():
    prefix = "[PENDIENTE] "

    def modify_first():
        nonlocal prefix
        # TODO: cambiar prefix a "[URGENTE] "
        prefix = "[URGENTE] "
        if tasks:
            tasks[0] = prefix + tasks[0]
        # TODO: si hay tareas, modificar la primera
        else:
            print("No hay tareas para modificar.")

    # TODO: aquí puedes simular algunas operaciones:
    while True:
        menu = input(" 1. Añadir tarea\n " \
        "2. Modificar tarea\n" \
        "3. Listar Tarea\n"
        "4. Eliminar tarea\n"
        "5. Salir\n " \
        "Elige una opción: ")

        # - añadir tareas
        if menu == "1":
            task=input("Escribe una tarea para añadir: ")
            add_task(task)

    # - llamar a modify_first()
        elif menu == "2":
            modify_first()
    # - listar tareas, etc.
        elif menu == "3":
            list_tasks()
        elif menu == "4":
            task=input("Escribe una tarea para eliminar: ")
            remove_task(task)
        elif menu == "5":
            break
        else:
            print("Opción no válida. Inténtalo de nuevo.")


def main():
    task_manager()


if __name__ == "__main__":
    main()
```

```python
def get_min_max(nums):
    """Devuelve una tupla (mínimo, máximo) sin usar min() ni max()."""
    # Comprobamos si la lista está vacía para evitar errores
    if not nums:
        return (None, None)

    # Inicializamos mínimo y máximo con el primer valor de la lista
    minimo = nums[0]
    maximo = nums[0]

    # Recorremos todos los números
    for n in nums:
        # Si encontramos un número menor que el actual mínimo, lo actualizamos
        if n < minimo:
            minimo = n
        # Si encontramos un número mayor que el actual máximo, lo actualizamos
        if n > maximo:
            maximo = n

    # Devolvemos los dos valores en una tupla
    return (minimo, maximo)


def get_average(nums):
    """Devuelve la media aritmética (float)."""
    # Si la lista está vacía, devolvemos 0.0 para evitar división por cero
    if not nums:
        return 0.0

    # Sumamos todos los elementos
    total = 0
    for n in nums:
        total += n

    # Calculamos la media dividiendo entre el número de elementos
    return total / len(nums)


def count_occurrences(nums, value):
    """Devuelve cuántas veces aparece value en nums (sin usar count())."""
    count = 0
    for n in nums:
        if n == value:
            count += 1
    return count


def main():
    # Pedimos al usuario una lista de números separados por espacios
    num = input("Introduce una lista de números separados por espacios: ")

    # Convertimos la cadena en una lista de enteros
    nums = [int(n) for n in num.split()]

    # Pedimos el valor a buscar
    value = int(input("Introduce el valor a buscar en la lista: "))

    # Llamamos a las funciones auxiliares
    minimo, maximo = get_min_max(nums)
    media = get_average(nums)
    ocurrencias = count_occurrences(nums, value)

    # Mostramos los resultados
    print(nums)
    print(f"\nMínimo: {minimo}")
    print(f"Máximo: {maximo}")
    print(f"Media: {media}")
    if ocurrencias == 1:
        print(f"El valor {value} aparece {ocurrencias} vez en la lista.")
    else:
        print(f"El valor {value} aparece {ocurrencias} veces en la lista.")


if __name__ == "__main__":
    main()

```

```python
def analyze_text(text):
    """
    Devuelve un diccionario con:
    - "lower": versión en minúsculas
    - "words": lista de palabras
    - "num_words": número de palabras
    - "longest": palabra más larga
    """
    # TODO: implementar usando lower(), split(), len(), etc.
    resultado = {}
    resultado["lower"] = text.lower()
    words = text.split()
    resultado["words"] = words
    resultado["num_words"] = len(words)
    resultado["longest"] = max(words, key=len) if words else ""
    return resultado


def main():
    # TODO: pedir frase al usuario
    frase = input("Introduce una frase: ")  
    # TODO: llamar a analyze_text
    resultados = analyze_text(frase)

    # TODO: mostrar resultados campo a campo
    print("Versión en minúsculas:", resultados["lower"])
    print("Lista de palabras:", resultados["words"])
    print("Número de palabras:", resultados["num_words"])
    print("Palabra más larga:", resultados["longest"])       


if __name__ == "__main__":
    main()
```

```python
def calculator(a, b, op):
    """Devuelve el resultado de aplicar op sobre a y b usando match-case."""
    match op:
        case "+":
            return a + b
        case "-":
            return a - b
        case "*":
            return a * b
        case "/":
            if b != 0:
                return a / b
            else:
                return "Error: División por cero"
        case _:
            return "Error: Operación no válida"


print(calculator(5, 2, "+"))  # 7
print(calculator(5, 2, "/"))  # 2.5
```

```python
def analyze_numbers(nums):
    sum = 0
    for i in nums:
        if i < 0:
            continue
        elif i == 0:
            break
        sum += i
    return sum
    
print(analyze_numbers([1, 2, -3, 4, 0, 5]))  # 7
```

```python
def location(x,y):
    match (x, y):
        case (0, 0):
            return "Origen"
        case (0, _):
            return "Eje Y"
        case (_, 0):
            return "Eje X"
        case _:
            return "Punto general"





print(location(0, 5))   # Eje Y
print(location(3, 0))   # Eje X
print(location(0, 0))   # Origen
print(location(2, 3))   # Punto general
```
